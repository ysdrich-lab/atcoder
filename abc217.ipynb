# A - Lexicographic Order 
https://atcoder.jp/contests/abc217/tasks/abc217_a
## 問題文

> 異なる二つの文字列 $S,T$ が与えられます。  
> $S$ が $T$ よりも辞書順で小さい場合は Yes を、大きい場合は No を出力してください。

## 制約

> $S,T$ は英小文字からなる長さ $1$ 以上 $10$ 以下の相異なる文字列である。

## 入力

> $S$&emsp;$T$

## 要点
* そのまま比較してあげて良し。
* $S$,$T$は異なる文字列のため、等しい状態を気にする必要もない。

```python
S,T = input().split()

if S < T:
    print('Yes')
else:
    print('No')
```

# B - AtCoder Quiz 
https://atcoder.jp/contests/abc217/tasks/abc217_b
## 問題文

> AtCoderでは現在、 ABC, ARC, AGC, AHC の 4 つのコンテストが定期的に開催されています。  
> AtCoderで現在定期的に開催されているコンテストは $S_1$,$S_2$,$S_3$とあと 1 つは何ですか?
## 制約

> * $S_1$,$S_2$,$S_3$はそれぞれ、 ABC , ARC , AGC , AHC のいずれかである。
> * $S_1$,$S_2$,$S_3$は相異なる。


## 入力

> $S_1$  
> $S_2$  
> $S_3$

## 要点
* 最初からコンテストの種類が決まっている
    * コンテスト４種類をリストとして持っておき、登場しなかった１つを抽出する

```python
#before
S1 = input()
S2 = input()
S3 = input()
 
l = ['ABC','ARC','AGC','AHC']

l.remove(S1)
l.remove(S2)
l.remove(S3)
 
print(l[0])
```

```python
# after
S1 = input()
S2 = input()
S3 = input()

input_l = [S1, S2, S3]
l = ['ABC', 'ARC', 'AGC', 'AHC']

for i in input_l: # ループで回したほうがお洒落ですね
    l.remove(i)

print(l[0])
```

# D - Cutting Woods 
https://atcoder.jp/contests/abc217/tasks/abc217_d
## 問題文
> 長さ L メートルの直線状の木材があります。  
> $x=1,2,…,L−1$に対して、木材の左端から$x$メートルの地点には目印として線$x$が引かれています。  
> $Q$個のクエリが与えられます。$i$番目のクエリは数の組$(c_i,x_i)$によって表されます。  
> 以下の説明に従ってクエリを$i$の昇順に処理してください。
> * $c_i=1$のとき:線$x_i$がある地点で木材を$2$つに切る。  
> * $c_i=2$のとき:線$x_i$を含む木材を選び、その長さを出力する。  
> 
> ただし$c_i=1,2$の両方に対して、線$x_i$はクエリを処理する時点で切られていないことが保証されます。


## 制約
> * $1 \leq L \leq 10^9$  
> * $0 \leq Q \leq 2\times 10^5$  
> * $c_i = 1,2(1 \leq i \leq Q)$
> * $1 \leq x_i \leq L-1(1 \leq i \leq Q)$
> * 全ての$i(1 \leq i \leq Q)$に対して次が成り立つ：$1 \leq j < i$かつ$(c_j,x_j) = (1,x_i)$を満たす$j$は存在しない
> * 入力は全て整数

## 入力
> $L$&emsp;$Q$  
> $c_1$&emsp;$x_1$  
> $c_2$&emsp;$x_2$  
> $\vdots$  
> $c_Q$&emsp;$x_Q$  


## 要点
### 順序付き集合
* 2番目のクエリを処理する際に必要な情報は、「$x$を超えて、最小の値」と「$x$未満で、最大の値」の２つの要素が高速に取得できれば良い。
* 作業手順としては以下の通り。
    * 集合$S$に$x$を追加
    * $S$に属する$x$より大きい要素のうち最小の要素を取得
    * $S$に属する$x$より小さい要素のうち最大の要素を取得
        * 順序付きの集合があれば解決ですね
        
Pythonには**順序付き集合が存在しない**。これは困った。

* 対処法１：平衝二分探索木
* 対処法２：Binary-Tree(二分トライ木)
* 対処法３：先に木を全て切断した上たいでUNION-FINDで後ろから考える（分割した木を結合する）

今の自分の技術、知識では対応できないため非推奨解法としてArray型で押し切る。

```python
import bisect
import array
 
L, Q = map(int, input().split())
 
cutted_l = array.array("i", [0, L])
 
for i in range(Q):
    c, x = map(int, input().split())
 
    if c == 1:
        bisect.insort(cutted_l, x)
    else:
        right_index = bisect.bisect(cutted_l, x)
        length = cutted_l[right_index] - cutted_l[right_index - 1]
        print(length)
```

# E - Sorting Queries
https://atcoder.jp/contests/abc217/tasks/abc217_e
## 問題文

> 空の列$A$があります。クエリが$Q$個与えられるので、与えられた順番に処理してください。  
> クエリは次の$3$種類のいずれかです。  
> * 1 $x$ : $A$の最後尾に$x$を追加する。  
> * 2 : $A$の最初の要素を出力する。その後、その要素を削除する。このクエリが与えられるとき、$A$は空でないことが保証される。  
> * 3 : $A$を昇順にソートする。

## 制約
> * $1 \leq Q \leq 2\times 10^5$  
> * $0 \leq x \leq 10^9$  
> * クエリ$2$が与えられる時、$A$は空ではない
> * 入力は全て整数

## 入力
> $Q$  
> $query1$  
> $query2$  
> $\vdots$  
> $queryQ$  


## 要点
### 計算量
* 文字通り実装するとTLEする。各クエリの計算量は、
    * クエリ1:$O(1)$
    * クエリ2:$O(n)$
    * クエリ3:$O(n\log n)$  

  になるから。「クエリ１が$\frac{Q}{2}$回、クエリ３が$\frac{Q}{2}$なんかで$O(Q^2 \log Q)$の計算量がかかる。

### 計算量の改善

* 配列を、ソート済みの部分と、未ソートの部分に分けて考える。  
  すると、この問題は次のように言い換えられる。
> 配列$A_1$とソート済み配列$A_2$がある。以下のクエリを処理せよ
> * クエリ１:$A_1$の最後尾に$x$を追加　　
> * クエリ２：$A_2$が空でない場合は$A_2$の最初の要素を、空の場合は$A_1$の先頭の要素を出力して削除
> * クエリ３：$A_1$の要素を全て$A_2$に移して$A_2$をソート　　

* この状態でクエリ１（未ソート配列$A_1$の最後尾に$x$を追加）は、キューを使えば$O(1)$で処理することができる。
* 問題はクエリ３。
* ここで、クエリ２で必要とされている情報を考える。
    * $A_1,A_2$共に先頭の要素しか必要とされていない。
        * $A_2$（ソート済み配列）に保持される必要があるのは**最小の要素**だけということ
        * $A_2$は優先度付きキュー（heapq)でデータ管理すると計算量を減らせる
* 具体例として、
> キュー$Q_1$と、優先度付きキュー$Q_2$を用意して、以下のクエリを処理する
> * クエリ１:$Q_1$に$x$をpushする
> * クエリ２：$Q_2$が空かどうか判断する
>   * 空の場合は、$Q_1$の先頭の要素を出力してpop  
>   * 空でない場合は、$Q_2$の最小の要素を出力してpop
> * クエリ３：$Q_1$の要素を全て$Q_2$に移す

* 計算量解析は少し複雑。
* ならし計算量とか、償却解析とか、償却計算量とか。
* 結果として計算量は$O(Q \log Q)$で解ける

```python
import heapq
from collections import deque

# 入力部
Q = int(input())

que = deque() # 未ソートの値を処理するためのキュー
sorted = [] # ソート済みの値を処理するためのheapq

# クエリ処理部
for i in range(Q):
    tmp = list(map(int, input().split()))

    if len(tmp) == 2:　# クエリの整形
        c = tmp[0]
        x = tmp[1]
    else:
        c = tmp[0]

    if c == 1: # クエリ１の場合、キューにxをpush
        que.append(x)
    elif c == 2: # クエリ２の場合、優先度付きキューが空か否かの判断
        if len(sorted) == 0: # 空の場合、キューの先頭をpop
            print(que.popleft())
        else: # 空でない場合、優先度付きキューの最小の要素をpop
            print(heapq.heappop(sorted))
    else: # クエリ３の場合、キューの要素を全て優先度付きキューに移す作業
        while que:
            heapq.heappush(sorted, que.popleft())
```
